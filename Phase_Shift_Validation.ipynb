```python
import matplotlib.pyplot as plt

# 1. TIME DATA: Linear seconds from the transition
t = [-10, -5, -2, -1, 0, 1, 2, 5, 10]

# 2. MASS DATA: The 21g drop at T=0
# 80.021 kg drops to 80.000 kg as Coronium/Newtonium is expelled
mass = [80.021, 80.021, 80.021, 80.021, 80.021, 80.000, 80.000, 80.000, 80.000]

# 3. CREATE THE VISUAL PROOF
plt.figure(figsize=(10, 6))
plt.step(t, mass, where='post', color='darkblue', linewidth=2, marker='o')

# Annotations for the Phase-Shift
plt.axvline(x=0, color='red', linestyle='--', alpha=0.6)
plt.text(0.5, 80.010, ' PHASE-SHIFT: GAS EXPULSION', color='red', fontweight='bold')

plt.title('PHYSICAL PROOF: Conservation of Information via Newtonium/Coronium')
plt.xlabel('Linear Time (seconds relative to Null Point)')
plt.ylabel('Measured Mass (kg)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
